# Data Visualization

## Plotting Histogram from 2001 to 2022

```python
cols = [str(i) for i in range(2001, 2023)]
for i in cols:
    plt.hist(df[i], color="#0099DD", bins=5)
    plt.xlabel(i)
    plt.show()
    
# Hid cell output for better view
```

## Plotting Barplot for each year

```python
# Extract the years (columns) from the DataFrame
years = df.columns[1:]  # Skip the first two columns ('Country Name')

# Sum the values for each year across all countries
total_values = df[years].sum()

# Plotting the bar chart
plt.figure(figsize=(15, 15))
plt.bar(years, total_values, color="#222200")
plt.xlabel("Year", size=20)
plt.ylabel("Total Values")
plt.title("Total Values per Year", size=20)
plt.show()
```

As we can see with the passage of time (from 2001 to 2022) the world population increases rapidly by about 30.5%

```python
p2022 = df["2022"].sum()
p2001 = df["2001"].sum()
print("Total Values in 2022: ", p2022)
print("Total Values in 2001: ", p2001)
print('Increase in Population from 2001 to 2022: ', (p2022 - p2001)/p2001 * 100, '%')
```

## Least Populated Countries in the years

```python
least_populated_countries = []
for year in range(2001, 2023):
    smallest_value = df[df[str(year)] == df[str(year)].min()]
    least_populated_countries.append(
        (
            year,
            smallest_value["Country Name"].iloc[0],
            smallest_value[str(year)].iloc[0],
        )
    )

data_least_populated = list(zip(*least_populated_countries))
plt.figure(figsize=(15, 15))
bars = plt.bar(data_least_populated[0], data_least_populated[2], color="#222200")
plt.xlabel("Years")
plt.ylabel("Population")
plt.title("Least Populated Countries in the World")

for i,country in enumerate(data_least_populated[1]):
    plt.text(
        bars[i].get_x() + bars[i].get_width() / 2,
        bars[i].get_height(),
        country,
        ha="center",
        va="bottom",
    )
plt.show()
```

As we can see the least populated country from 2001 till 2007 is Tuvalu which was later taken by Nauru and attained back by Tuvalu.
